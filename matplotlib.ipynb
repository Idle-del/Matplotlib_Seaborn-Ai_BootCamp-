```python
import pandas as pd
import numpy as np

np.random.seed(42)

months = pd.date_range('2024-01-01', '2024-12-31', freq='D')
categories = ['Electronics', 'Clothing', 'Groceries', 'Furniture', 'Toys']

data = []

for month in months:
    for category in categories:
        data.append({
            'month': month,
            'category': category,
            'sales': np.random.randint(5000, 20000),
            'profit': np.random.uniform(0.1, 0.4)
        })
ecom_df = pd.DataFrame(data)
print(ecom_df)
```

```python
np.random.seed(24)

employees = 200
employees_data = ({
    'employee_id': range(1, employees + 1),
    'department': np.random.choice(['Sales', 'Marketing', 'IT', 'HR', 'Finance'], employees),
    'salary': np.random.randint(40000, 120000, employees),
    'age': np.random.randint(22, 60, employees)
})

employees_data['experience_years(this company)'] = [np.random.randint(0, max(1, age - 21)) for age in employees_data['age']]

employees_df = pd.DataFrame(employees_data)
print(employees_df.head().to_string(index=False))
```

```python
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = (10, 6)
plt.rcParams['font.size'] = 12
plt.rcParams['axes.grid'] = True
plt.rcParams['grid.alpha'] = 0.3
```

```python
## Assignment 1: Create a multi panel dashboard
import seaborn as sns

sns.set(style="whitegrid")

np.random.seed(42)

# sales over time
dates = pd.date_range(start='2025-01-01', end='2025-10-01', freq='D')
sales_over_time = np.random.randint(50, 500, size=len(dates))

# categories distribution
categories = ['Electronics', 'Clothing', 'Home', 'Books', 'Toys']
category_sales = np.random.randint(1000, 5000, size=len(categories))

# regional comparison
regions = ['North', 'South', 'East', 'West']
regional_sales = np.random.randint(2000, 7000, size=len(regions))

# top performing products
products = ['Laptop', 'Smartphone', 'Shoes', 'Book', 'Blender']
product_sales = np.random.randint(500, 3000, size=len(products))

fig, axes = plt.subplots(2, 2, figsize=(15, 10))
fig.suptitle('Comprehensive Sales Dashboard', fontsize=20, fontweight='bold')

## Sales Over Time
sns.lineplot(x=dates, y=sales_over_time, ax=axes[0,0], color='b')
axes[0,0].set_title('Sales Over Time', fontsize=14, fontweight='bold')
axes[0,0].set_xlabel('Date')
axes[0,0].set_ylabel('Sales')
axes[0,0].tick_params(axis='x', rotation=45)

## Categories Distribution
sns.barplot(x=categories, y=category_sales, ax=axes[0,1])
axes[0,1].set_title('Category Distribution', fontsize=14, fontweight='bold')
axes[0,1].set_xlabel('Category')
axes[0,1].set_ylabel('Sales')

## Top Performing Products
sns.barplot(y=products, x=product_sales, ax=axes[1,0])
axes[1,0].set_title('Top Performing Products', fontsize=14, fontweight='bold')
axes[1,0].set_xlabel('Sales')
axes[1,0].set_ylabel('Product')

## Regional Sales Comparison
axes[1,1].pie(regional_sales, labels=regions, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
axes[1,1].set_title('Regional Sales Comparison', fontsize=14, fontweight='bold')
axes[1,1].set_xlabel('Region')

plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

```python
np.random.seed(42)

data = pd.DataFrame({
    'Age': np.random.randint(18, 70, 200),
    'Annual_Income': np.random.randint(20000, 120000, 200),
    'Spending_Score': np.random.randint(1, 100, 200),
    'Years_With_Company': np.random.randint(0, 20, 200),
    'Purchases_Last_Year': np.random.randint(0, 50, 200),
    'Online_Visits': np.random.randint(0, 200, 200),
    'Satisfaction_Score': np.random.randint(1, 10, 200),
    'Support_Calls': np.random.randint(0, 10, 200),
    'Region': np.random.choice(['North', 'South', 'East', 'West'], 200)
})

print(data.head().to_string(index=False))
```

```python
## Assignment 2: Statistical Analysis Visualization
sns.set(style="whitegrid")

fig, axes = plt.subplots(3, 3, figsize=(20, 15))
fig.suptitle('Comprehensive Customer Data Analysis', fontsize=20)

axes = axes.flatten()

# Age Distribution
sns.histplot(data['Age'], kde=True, color='skyblue', ax=axes[0])
axes[0].set_title('Age Distribution')
axes[0].axvline(data['Age'].mean(), color='red', linestyle='--', label=f"Mean: {data['Age'].mean():.1f}")
axes[0].axvline(data['Age'].median(), color='green', linestyle='-', label=f"Median: {data['Age'].median():.1f}")
axes[0].legend()

# Annual Income Distribution
sns.histplot(data['Annual_Income'], kde=True, color='lightgreen', ax=axes[1])
axes[1].set_title('Annual Income Distribution')
axes[1].axvline(data['Annual_Income'].mean(), color='red', linestyle='--', label=f"Mean: {data['Annual_Income'].mean():.1f}")
axes[1].axvline(data['Annual_Income'].median(), color='green', linestyle='-', label=f"Median: {data['Annual_Income'].median():.1f}")
axes[1].legend()

# Spending Score Distribution
sns.histplot(data['Spending_Score'], kde=True, color='salmon', ax=axes[2])
axes[2].set_title('Spending Score Distribution')
axes[2].axvline(data['Spending_Score'].mean(), color='red', linestyle='--', label=f"Mean: {data['Spending_Score'].mean():.1f}")
axes[2].axvline(data['Spending_Score'].median(), color='green', linestyle='-', label=f"Median: {data['Spending_Score'].median():.1f}")
axes[2].legend()

# Years With Company Distribution
sns.histplot(data['Years_With_Company'], color='lightgreen', ax=axes[3])
axes[3].set_title('Years With Company Distribution')
axes[3].axvline(data['Years_With_Company'].mean(), color='red', linestyle='--', label=f"Mean: {data['Years_With_Company'].mean():.1f}")
axes[3].axvline(data['Years_With_Company'].median(), color='green', linestyle='-', label=f"Median: {data['Years_With_Company'].median():.1f}")
axes[3].legend()

# Purchases Last Year Distribution
sns.histplot(data['Purchases_Last_Year'], kde=True, color='lightblue', ax=axes[4])
axes[4].set_title('Purchases Last Year Distribution')

# Online Visits Distribution
sns.histplot(data['Online_Visits'],  color='cyan', ax=axes[5])
axes[5].set_title('Online Visits Distribution')
axes[5].axvline(data['Online_Visits'].mean(), color='red', linestyle='--', label=f"Mean: {data['Online_Visits'].mean():.1f}")
axes[5].axvline(data['Online_Visits'].median(), color='green', linestyle='-', label=f"Median: {data['Online_Visits'].median():.1f}")
axes[5].legend()

# Satisfaction Score Distribution
sns.scatterplot(x='Spending_Score', y='Satisfaction_Score', data=data, hue='Region', ax=axes[6])
axes[6].set_title('Satisfaction Score Distribution')

# Support Calls Distribution
sns.boxplot(x='Region', y='Support_Calls', data=data, color='lightcoral', ax=axes[7])
axes[7].set_title('Support Calls Boxplot')

corr = data.select_dtypes(include=np.number).corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', ax=axes[8])
axes[8].set_title('Correlation Heatmap')

plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```
