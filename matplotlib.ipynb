```python
import pandas as pd
import numpy as np

np.random.seed(42)

months = pd.date_range('2024-01-01', '2024-12-31', freq='D')
categories = ['Electronics', 'Clothing', 'Groceries', 'Furniture', 'Toys']

data = []

for month in months:
    for category in categories:
        data.append({
            'month': month,
            'category': category,
            'sales': np.random.randint(5000, 20000),
            'profit': np.random.uniform(0.1, 0.4)
        })
ecom_df = pd.DataFrame(data)
print(ecom_df)
```

```python
np.random.seed(24)

employees = 200
employees_data = ({
    'employee_id': range(1, employees + 1),
    'department': np.random.choice(['Sales', 'Marketing', 'IT', 'HR', 'Finance'], employees),
    'salary': np.random.randint(40000, 120000, employees),
    'age': np.random.randint(22, 60, employees)
})

employees_data['experience_years(this company)'] = [np.random.randint(0, max(1, age - 21)) for age in employees_data['age']]

employees_df = pd.DataFrame(employees_data)
print(employees_df.to_string(index=False))
```

```python
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = (10, 6)
plt.rcParams['font.size'] = 12
plt.rcParams['axes.grid'] = True
plt.rcParams['grid.alpha'] = 0.3
```

```python
## Assignment 1: Create a multi panel dashboard
import seaborn as sns

sns.set(style="whitegrid")

np.random.seed(42)

# sales over time
dates = pd.date_range(start='2025-01-01', end='2025-10-01', freq='D')
sales_over_time = np.random.randint(50, 500, size=len(dates))

# categories distribution
categories = ['Electronics', 'Clothing', 'Home', 'Books', 'Toys']
category_sales = np.random.randint(1000, 5000, size=len(categories))

# regional comparison
regions = ['North', 'South', 'East', 'West']
regional_sales = np.random.randint(2000, 7000, size=len(regions))

# top performing products
products = ['Laptop', 'Smartphone', 'Shoes', 'Book', 'Blender']
product_sales = np.random.randint(500, 3000, size=len(products))

fig, axes = plt.subplots(2, 2, figsize=(15, 10))
fig.suptitle('Comprehensive Sales Dashboard', fontsize=20, fontweight='bold')

## Sales Over Time
sns.lineplot(x=dates, y=sales_over_time, ax=axes[0,0], color='b')
axes[0,0].set_title('Sales Over Time', fontsize=14, fontweight='bold')
axes[0,0].set_xlabel('Date')
axes[0,0].set_ylabel('Sales')
axes[0,0].tick_params(axis='x', rotation=45)

## Categories Distribution
sns.barplot(x=categories, y=category_sales, ax=axes[0,1])
axes[0,1].set_title('Category Distribution', fontsize=14, fontweight='bold')
axes[0,1].set_xlabel('Category')
axes[0,1].set_ylabel('Sales')

## Top Performing Products
sns.barplot(y=products, x=product_sales, ax=axes[1,0])
axes[1,0].set_title('Top Performing Products', fontsize=14, fontweight='bold')
axes[1,0].set_xlabel('Sales')
axes[1,0].set_ylabel('Product')

## Regional Sales Comparison
axes[1,1].pie(regional_sales, labels=regions, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
axes[1,1].set_title('Regional Sales Comparison', fontsize=14, fontweight='bold')
axes[1,1].set_xlabel('Region')

plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```
